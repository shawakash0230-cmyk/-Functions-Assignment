**1. What is the difference between a function and a method in Python?**

- In Python:

  - A function is a standalone block of code that performs a specific task. It's defined using def and can be called independently. Example: len(), print().

  - A method is a function that's associated with an object (like a string, list, etc.). It's called on an object using dot notation. Example: my_list.append(), my_str.upper().


Key difference: Methods are tied to objects; functions aren't.

```python
#Function: len()

my_list = [1, 2, 3]
print(len(my_list))  # Output: 3


#Method: append()

my_list = [1, 2, 3]
my_list.append(4)
print(my_list)  # Output: [1, 2, 3, 4]


#In the first case, len() is a function that takes an object (my_list) as an argument. In the second case, append() is a method called on the my_list object.
```

**3. What are the different ways to define and call a function in Python?**

- In Python, functions can be defined and called in several ways:

  1. Positional arguments: def func(a, b):
  2. Keyword arguments: func(a=1, b=2)
  3. Default arguments: def func(a, b=2):
  4. *Variable arguments (_args)_: def func(*args):
  5. *Keyword-only arguments (*kwargs): def func(**kwargs):
  6. Lambda functions: func = lambda x: x*2
  7. Nested functions: defining a function inside another function.

You can call functions using:

  - Positional arguments: func(1, 2)
  - Keyword arguments: func(a=1, b=2)
  - Mix of both: func(1, b=2)

```python
#1. Positional arguments

def add(a, b):
    return a + b

result = add(3, 5)
print(result)  # Output: 8


#2. Keyword arguments

def greet(name, msg):
    print(f"{msg}, {name}!")

greet(name="Alice", msg="Hello")
# Output: Hello, Alice!


#3. Default arguments

def greet(name = "Guest"):
    print(f"Hello, {name}!")

greet()  # Output: Hello, Guest!
greet("Alice")  # Output: Hello, Alice!


#*4. Variable arguments (_args)_

def sum_numbers(*args):
    return sum(args)

print(sum_numbers(1, 2, 3))  # Output: 6


#*5. Keyword-only arguments (*kwargs)

def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=25)
# Output:
# name: Alice
# age: 25


#6. Lambda functions

add = lambda a, b: a + b
result = add(3, 5)
print(result)  # Output: 8
```

**5. What are iterators in Python and how do they differ from iterables?**

- In Python:
  - Iterables: Objects that can be looped over (e.g., lists, tuples, strings). They implement the __iter__() method.
  - Iterators: Objects that keep track of their position in an iterable. They implement __iter__() and __next__() methods.

Key difference:
  - Iterables can be looped over multiple times.
  - Iterators can be looped over only once; they "exhaust" after use.

```python
my_list = [1, 2, 3]  # iterable
my_iter = iter(my_list)  # iterator

print(next(my_iter))
print(next(my_iter))
print(next(my_iter))
```

**6. Explain the concept of generators in Python and how they are defined.**

- In Python, generators are a type of iterable, like lists or tuples. They're defined using functions with the yield keyword instead of return.

Key characteristics:
  - Generators produce values on-the-fly, one at a time.
  - They remember their state between calls.
  - They're memory-efficient, as they don't store all values at once.

Generators are often used for:
  - Handling large datasets
  - Creating infinite sequences
  - Simplifying code with iterative logic

```python
def my_generator():
    yield 1
    yield 2
    yield 3

for value in my_generator():
    print(value)
```

**7. What are the advantages of using generators over regular functions?**

- Advantages of generators over regular functions:
  - Memory efficiency: Generators produce values on-the-fly, reducing memory usage.
  - Lazy evaluation: Values are generated only when needed.
  - Handling large datasets: Generators can handle infinite or large sequences.
  - Improved performance: No need to compute and store all values upfront.
  - Simplified code: Generators can simplify iterative logic.

```python
#Regular Function (returns a list)

def get_squares(n):
    result = []
    for i in range(n):
        result.append(i * i)
    return result

print(get_squares(5))
```

```python
#Generator Function (yields values)

def generate_squares(n):
    for i in range(n):
        yield i * i

for square in generate_squares(5):
    print(square)
```

**9. Explain the purpose and usage of the `map()` function in Python.**

- In Python, the map() function:
  - Applies a given function to each item of an iterable (e.g., list, tuple).
  - Returns a map object (an iterator) with the results.

Syntax: map(function, iterable)

Common use cases:
  - Data transformation
  - Applying a function to multiple inputs
  - Simplifying code with iterative logic

```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x**2, numbers)
print(list(squared))
```

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

- In Python:
  - map(function, iterable): Applies a function to each item, returning a new iterable with transformed values.
  - filter(function, iterable): Returns an iterable with items for which the function returns True.
  - reduce(function, iterable) (from functools module): Applies a function cumulatively to items, reducing the iterable to a single value.

```python
#Map

numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

```python
#Reduce

from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

```python
#Filter

numbers = [1, 2, 3, 4, 5]
even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)
```

**1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**

```python
def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

numbers = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(numbers)
print(result)
```

**4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(n):
    """Check if a number is prime between 1 and 200."""
    if n < 2 or n > 200:
        return False  # Not prime or out of range
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number.")
```

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

**6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(max_exponent):
    """Yield powers of 2 from 2^0 up to 2^max_exponent."""
    for exp in range(max_exponent + 1):
        yield 2 ** exp
for value in powers_of_two(10):
    print(value)
```

**7. Implement a generator function that reads a file line by line and yields each line as a string.**

```python
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
filename = 'large_file.txt'
for line in read_file_lines(filename):
    print(line)
```

**9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
# List of temperatures in Celsius
celsius_temps = [0, 10, 20, 30, 40, 100]

# Conversion formula: F = C * 9/5 + 32
fahrenheit_temps = list(map(lambda c: c * 9/5 + 32, celsius_temps))

# Output the result
print("Celsius:", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)
```

**10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda ch: ch not in vowels, input_str))

text = "Hello, World!"
no_vowels = remove_vowels(text)
print("Original:", text)
print("Without vowels:", no_vowels)
```
